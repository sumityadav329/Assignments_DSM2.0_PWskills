**Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.**

**A1. Matplotlib:**
- Matplotlib is a 2D plotting library for the Python programming language. It enables the creation of static, animated, and interactive visualizations in Python. Matplotlib is widely used for creating high-quality charts, plots, and figures, making it an essential tool for data visualization and analysis.

**Why is it used?**
- Matplotlib is used for various purposes, including:

1. **Data Visualization:** Matplotlib allows users to create a wide range of static, animated, and interactive plots, helping in the visual representation of data.

2. **Publication-Quality Plots:** It provides a high level of control over plot aesthetics, allowing users to create publication-quality figures and customize plots extensively.

3. **Integration with Jupyter Notebooks:** Matplotlib integrates seamlessly with Jupyter Notebooks, making it a preferred choice for creating visualizations in a notebook environment.

4. **Wide Range of Plot Types:** Matplotlib supports a variety of plot types, enabling users to visualize data in different ways for better understanding and analysis.

5. **Community Support:** Being a widely used library, Matplotlib has a large and active community, ensuring continuous development, support, and availability of resources.

**Five Plots that can be plotted using Pyplot module of Matplotlib:**

1. **Line Plot:**

````python

import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.figure(figsize=(6,4))
plt.plot(x, y)
plt.show()
````

2. **Scatter Plot:**
   

````python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.figure(figsize=(6,4))
plt.scatter(x, y)
plt.show()
````

3. **Bar Chart:**

````python
categories = ['A', 'B', 'C', 'D', 'E']
values = [3, 7, 1, 5, 10]
plt.figure(figsize=(6,4))
plt.bar(categories, values)
plt.show()
````

4. **Histogram:**

````python

data = [2, 3, 3, 4, 4, 4, 5, 5, 6, 7]
plt.figure(figsize=(6,4))
plt.hist(data, bins=10)
plt.show()
````

5. **Pie Chart:**

````python

sizes = [30, 25, 20, 15, 10]
labels = ['A', 'B', 'C', 'D', 'E']
plt.figure(figsize=(6,4))
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.show()
````

**Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.**
```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```
**Note: Also add title, xlabel, and ylabel to the plot.**

**A2. A scatter plot is a type of data visualization that displays individual data points on a two-dimensional graph. Each point on the graph represents the values of two different variables. Scatter plots are particularly useful for visualizing the relationship between two continuous variables and identifying patterns or trends in the data.**

In a scatter plot:

- Each point on the graph represents a single data observation.
- The x-axis and y-axis represent the two variables being compared.
- The position of each point is determined by the values of the two variables for that specific data point.

Scatter plots are commonly used to:

1. **Identify Patterns:** Scatter plots help in identifying patterns or relationships between two variables. For example, you can visually assess whether there is a positive, negative, or no correlation between the variables.

2. **Outlier Detection:** Outliers, or data points that deviate significantly from the general pattern, can be easily spotted in a scatter plot.

3. **Correlation Analysis:** Scatter plots can provide insights into the strength and direction of the correlation between two variables. Correlation can be positive (both variables increase together), negative (one variable increases while the other decreases), or neutral (no apparent relationship).

Here's a simple example of creating a scatter plot using Matplotlib in Python:

````python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.figure(figsize=(8,6))
plt.scatter(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('X vs Y plot')
plt.show()
````

**Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:**
```python
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

**A3. 
The `subplot()` function in Matplotlib is used to create multiple plots (subplots) within the same figure. This function is particularly useful when you want to compare different visualizations or datasets side by side, facilitating a better understanding of the relationships and patterns in the data. The `subplot()` function allows you to create a grid of subplots and specify the position of each subplot within that grid.**

````python
#in given data x is same for every line that why renaming y as y1 , y2, y3, y4
x = np.array([0, 1, 2, 3, 4, 5]) 

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

#creating plot 
plt.subplot(2,2,1)
plt.plot(x,y1, color = 'blue', marker= 'o')
plt.title('Line plot 1')

plt.subplot(2,2,2)
plt.plot(x,y2, color = 'green', marker= 's')
plt.title('Line plot 2')

plt.subplot(2,2,3)
plt.plot(x,y3, color = 'red', marker= '^')
plt.title('Line plot 3')

plt.subplot(2,2,4)
plt.plot(x,y4, color = 'orange', marker= 'd')
plt.title('Line plot 4')

#adjusting layout for better spacing
plt.tight_layout()

#showing the plot
plt.show()
````

**Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.**
```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

**A4. Bar Plot:**
- A bar plot is a data visualization that presents categorical data with rectangular bars. The lengths of the bars are proportional to the values they represent. Bar plots are useful for comparing the magnitudes of different categories or groups. They provide a clear visual representation of the data and make it easy to identify trends or variations.

**Why is it used:**
- **Comparison:** Bar plots are effective for comparing the values of different categories or groups.
- **Categorical Data:** Suitable for displaying data with distinct categories.
- **Easy Interpretation:** The length of each bar directly corresponds to the quantity it represents, making it easy to interpret the data.

Now, let's create a bar plot and a horizontal bar plot using the provided data:


````python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar Plot
plt.figure(figsize=(6, 4))
plt.bar(company, profit, color=['blue', 'green', 'red', 'purple'])
plt.xlabel('Company')
plt.ylabel('Profit (in billions)')
plt.title('Profit of Companies')
plt.show()

````

````python
# Horizontal Bar Plot
plt.figure(figsize=(6,4))
plt.barh(company, profit, color=['blue', 'green', 'red', 'purple'])
plt.xlabel('Profit (in billions)')
plt.ylabel('Company')
plt.title('Profit of Companies (Horizontal)')
plt.show()
````

**Q5: What is a box plot? Why is it used? Using the following data plot a box plot.**
```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

**A5. Box Plot:**
- A box plot, also known as a box-and-whisker plot, is a statistical visualization that displays the distribution and spread of a dataset. It provides a summary of key statistical measures, including the median, quartiles, and potential outliers. Box plots are particularly useful for comparing the distributions of different datasets and identifying patterns within the data.

**Why is it used:**
- **Summary of Distribution:** Box plots provide a concise summary of the distribution, including the median, quartiles, and the range of the data.
- **Outlier Detection:** Outliers, or data points significantly different from the majority, can be easily identified in a box plot.
- **Comparison:** Box plots are useful for comparing the spread and central tendency of multiple datasets.

Now, let's create a box plot using the provided data:



````python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

#creating box plot
plt.figure(figsize= (8,6))
plt.boxplot([box1, box2 ], labels = ['Box 1','Box 2'])
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box plot of Box 1 and Box 2')
plt.show()
                                    
````
