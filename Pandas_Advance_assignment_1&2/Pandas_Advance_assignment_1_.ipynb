DataFrame with a new index that starts from 1 and increments by 2 for each row.**

````python
# to re-index the df creating a new function
def reindex_dataframe(df):
    #creating a new index starting from 1 and incrementing by 2 for each row
    new_index = range(1,2*len(df) + 1,2)
    
    #setting the new index for the Dataframe
    reindx_df = df.set_index(pd.Index(new_index))
    
    return reindx_df

reindx_df = reindex_dataframe(df)   
````

**Q5. How are DataFrame.size() and DataFrame.shape() different?**
\
**A5.**
\
- In Pandas, `DataFrame.size` and `DataFrame.shape` are two different attributes that provide different information about the structure of a DataFrame.


1. **`DataFrame.size`:**
   - `DataFrame.size` returns the total number of elements in the DataFrame.
   - It calculates the size by multiplying the number of rows by the number of columns.
   - The result is a single integer representing the total number of cells in the DataFrame.

````python
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
size_result = df.size
print("DataFrame size:", size_result)
````

2. **`DataFrame.shape`:**
   - `DataFrame.shape` returns a tuple representing the dimensions of the DataFrame.
   - The tuple contains two elements: the number of rows and the number of columns.
   - It provides a convenient way to check the shape of the DataFrame without calculating the total number of elements.

````python
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
shape_result = df.shape
print("DataFrame shape:", shape_result)
````

\
Write a Python function that selects
all rows where the value in column 'A' is greater than 5 and the value in column 'B' is less than 10. 
\
The function should return a new DataFrame that contains only the selected rows.
For example, if df contains the following values:**
|  |A|B|C|
|--|-|---|--|
|0 |3| 5 | 1|
|1 |8| 2 | 7|
|2 |6| 9 | 4|
|3 |2| 3 | 5|
|4 |9| 1 | 2|

**Your function should select the following rows:**

|A| B|C|
|-|-|-|
|1|8| 2 |7|
|4|9|1 |2|

- The function should return a new DataFrame that contains only the selected rows.

**A 8.**

````python
# creating df as question
df = pd.DataFrame({'A':[3,8,6,2,9],
                  'B':[5,2,9,3,1],
                  'C':[1,7,4,5,2]})
df
````

````python
def select_rows(df):
   
     #Check if 'A' and 'B' column exists in the DataFrame
    if 'A'  not in df.columns or 'B' not in df.columns:
        print("Error: 'A' and 'B' column not found in DataFrame.")
        return
    
    # Select rows where 'A' is greater than 5 and 'B' is less than 10
    selected_rows = df[(df['A']>5) & (df['B']<10)]
    return selected_rows

# Call the function to select rows based on conditions
selected_df = select_rows(df)     
````

**Q9. Given a Pandas DataFrame df with a column 'Values', write a Python function to calculate the mean,
median, and standard deviation of the values in the 'Values' column.**
\
**A9.**

````python
df = pd.DataFrame({'Values': [10,20,30,40,50]})
df
````

````python
def calculate_mean_med_std(df):
    # check 'Values' column really exists in df or not
    if "Values" not in df.columns:
        print("Error: 'Values' not found in Dataframe ")
        return None
    mn = df['Values'].mean()
    med = df['Values'].median()
    std = df['Values'].std()
    
    return mn,med,std
calculated_values = calculate_mean_med_std(df)
````

````python
calculated_df = pd.DataFrame({"Stats": ["mean","median","std"],
                             "values": calculated_values})
calculated_df
````

**OR We can make this code shorter using `.agg()`** 

````python
def calculate_mean_med_std(df):
    if 'Values' not in df.columns:
        print("Error: 'Values' column not found in DataFrame.")
        return None
    
    calculated_values = df['Values'].agg(['mean', 'median', 'std'])
    
    return pd.DataFrame({'Stats': calculated_values.index, 'Value': calculated_values.values})

calculated_df = calculate_mean_med_std(df)

calculated_df
````

**Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', 
\
write a Python function to create a new column 'MovingAverage' that contains 
\
the moving average of the sales for the past 7 days for each row in the DataFrame. 
\
The moving average should be calculated using a window of size 7 and
should include the current day.
\
A10.**

````python
# Example as question asks
data = {'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05'],
        'Sales': [10, 15, 12, 18, 20]}

df = pd.DataFrame(data)
df
````

````python
def calculate_moving_average(df):
    # Check if 'Sales' and 'Date' columns exist in the DataFrame
    if 'Sales' not in df.columns or 'Date' not in df.columns:
        print("Error: 'Sales' or 'Date' column not found in DataFrame.")
        return None

    # Converting 'Date' column to datetime format
    df['Date'] = pd.to_datetime(df['Date'])

    # Sorting DataFrame by 'Date' in case it is not already sorted
    df.sort_values(by='Date', inplace=True)

    # Calculating the moving average using a window of size 7
    df['MovingAverage'] = df['Sales'].rolling(window=7, min_periods=1).mean()

    return df


# Calling the function 
df_with_ma = calculate_moving_average(df)


````

**Q11. You have a Pandas DataFrame df with a column 'Date'. Write a Python function that creates a new
column 'Weekday' in the DataFrame. The 'Weekday' column should contain the weekday name (e.g.
Monday, Tuesday) corresponding to each date in the 'Date' column.
\
For example, if df contains the following values:**

|  |  Date    |
|--|----------|
|0 |2023-01-01|
|1 |2023-01-02|
|2 |2023-01-03|
|3 |2023-01-04|
|4 |2023-01-05|

Your function should create the following DataFrame:


|  |    Date   | Weekday  |
|--|-----------|--------- |
|0 |2023-01-01 | Sunday   |
|1 |2023-01-02 |Monday    |
|2 |2023-01-03 |Tuesday   |
|3 |2023-01-04 | Wednesday|
|4 |2023-01-05 |Thursday  |

The function should return the modified DataFrame.

**A11.**


````python
# Example as given
data = {'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']}
df = pd.DataFrame(data)
df
````

````python
def add_weekday_column(df):
    # Check if 'Date' column exists in the DataFrame
    if 'Date' not in df.columns:
        print("Error: 'Date' column not found in DataFrame.")
        return None
    
    # Convert 'Date' column to datetime format
    df['Date'] = pd.to_datetime(df['Date'])

    # Create a new 'Weekday' column containing the weekday names
    df['Weekday'] = df['Date'].dt.strftime('%A')

    return df


df_with_weekday = add_weekday_column(df)
````

**Q12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, 
\
write a Python function to select all rows where the date is between '2023-01-01' and '2023-01-31'.
\
A12.**

````python
# Example 
data = {'Date': ['2023-01-01', '2023-01-15', '2023-01-20', '2023-02-05', '2023-02-10']}
df = pd.DataFrame(data)
df
````

````python
def select_rows_in_date_range(df):
    # Check if 'Date' column exists in the DataFrame
    if 'Date' not in df.columns:
        print("Error: 'Date' column not found in DataFrame.")
        return None
    
    # Convert 'Date' column to datetime format
    df['Date'] = pd.to_datetime(df['Date'])

    # Select rows where the date is between '2023-01-01' and '2023-01-31'
    selected_rows = df[(df['Date'] >= '2023-01-01') & (df['Date'] <= '2023-01-31')]

    return selected_rows

selected_df = select_rows_in_date_range(df)

selected_df
````
