### Q2. What is the difference between the functions loc and iloc in pandas.DataFrame?
### A2.

In Pandas, `loc` and `iloc` are two different methods used for indexing and selecting data in a DataFrame.

 `loc` is label-based, and we use it when we want to select data based on row and column labels. `iloc` is integer-location based, and we use it when we want to select data based on integer indices.

1. **`loc` (label-based selection):**
   - `loc` is primarily label-based, meaning that you use column and index names to select data.
   - It uses labels or boolean arrays for selection.
   - You specify the row and column labels explicitly.
   - The syntax is `df.loc[row_label, column_label]`.

2. **`iloc` (integer-location based selection):**
   - `iloc` is primarily integer-location based, meaning that you use integer indices to select data.
   - It is useful when you want to select data by position rather than by label.
   - You specify the row and column indices using integers.
   - The syntax is `df.iloc[row_index, column_index]`.

````python
df1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]}, index=['a', 'b', 'c'])

# Using loc for label-based selection
result_loc = df1.loc['b', 'B']

# Using iloc for integer-location based selection
result_iloc = df1.iloc[1, 1]

print("Using loc:", result_loc)
print("Using iloc:", result_iloc)
````

### Q6. What do you understand about the windows function in pandas and list the types of windows functions?
### A6.

**In Pandas, the term "window functions" refers to a group of functions that perform calculations on a specified window of data in a time series or DataFrame. These functions are particularly useful for tasks involving rolling or expanding windows, where you want to calculate statistics over a specific subset of data points.**

````python
# first we need a dataframe we are using df1 as given for questions
columns = ['column_1', 'column_2', 'column_3', 'column_4', 'column_5', 'column_6']
indices = [1,2,3,4,5,6]
#Creating a dataframe:
df1 = pd.DataFrame(np.random.rand(6,6), columns = columns, index = indices)
df1
````

### Types of window functions in Pandas include:


1. **Rolling Functions:**
   - Rolling functions operate on a fixed-size window of data that moves along the time series or DataFrame.
   - These functions are useful for tasks such as calculating rolling averages, rolling sums, and other moving window statistics.
   - Example: `rolling_mean`, `rolling_sum`, `rolling_std`

````python
# Example of rolling mean calculation
df1['rolling_mean'] = df1['column_1'].rolling(window=3).mean()
df1
````

2. **Expanding Functions:**
   - Expanding functions consider all the data points from the start of the time series up to the current point.
   - These functions are useful for tasks where you want to calculate cumulative statistics.
   - Example: `expanding_mean`, `expanding_sum`, `expanding_std`

````python
# Example of expanding mean calculation
df1['expanding_mean'] = df1['column_2'].expanding().mean()
df1
````

3. **Exponential Moving Average (EMA):**
   - EMA is a weighted moving average that gives more weight to recent observations.
   - It is often used to smooth time series data.
   - Example: `ewm`  

````python
# Example of EMA calculation
df1['ema'] = df1['column_3'].ewm(span=2, adjust=False).mean()
df1
````

4. **Window Function with `apply`:**
   - You can use the `apply` method with a custom function to perform calculations on a rolling or expanding window.
   - This provides flexibility for more complex calculations.
   - Example: Custom function using `apply`

````python
# Example of custom function with rolling window
def custom_function(window_data):
# Perform custom calculations on the window_data
    return window_data.sum() / window_data.mean()

df1['custom_result'] = df1['column_4'].rolling(window=3).apply(custom_function)
df1
````

### Did you observe any difference in both the outputs? If so then explain it.


### A8.



````python
# function to calculate difference b/w time of dates
def calculate_time_difference(date1,date2):
    # convertin input dates to datetime objects 
    date1,date2 = pd.to_datetime(date1),pd.to_datetime(date2)
    # calculate the difference 
    time_difference = date1 - date2
    # extract days, hours, minutes from the time difference 
    days = time_difference.days
    hours, remainder = divmod(time_difference.seconds, 3600)
    minutes = remainder // 60
    return days, hours, minutes

#promting the user to input the dates
date_input1 = input("Enter the date in format(YYYY-MM-DD): ")
date_input2 = input("Enter the date in format(YYYY-MM-DD): ")

result_days,result_hours,result_minutes = calculate_time_difference(date_input1,date_input2)

print("\nTime Difference : ")
print(f"Days : {result_days}")
print(f"Hours : {result_hours}")   
print(f"Minutes : {result_minutes}")
````

````python
#function to convert specific column to categorical and display the sorte data
def convert_col_to_cat(file_path,column_name,category_order):
    
    try:
        df = pd.read_csv(file_path)
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found")
        return
    except pd.errors.EmptyDataError:
        print('File is empty')
        return
    
    df[column_name] = pd.Categorical(df[column_name], categories = category_order, ordered=True)
    
    # display the sorted data
    sorted_data = df.sort_values(by = [column_name])
    print("\nSorted Data:")
    print(sorted_data)
    
    
#promt the user to enter file path, column name,and category order
file_path = input("Enter the file path (CSV):")
column_name = input("Enter the column name to convert to categorical:")
category_order_str = input("Enter the category order (comma-separated):")

#conerting the comma-seperated category order to a list
category_order = [category.strip() for category in category_order_str.split(',')]

#calling the function to convert and display the sorted data
convert_col_to_cat(file_path, column_name, category_order)
````

Q10. Write a Python program that reads a CSV file containing sales data for different products and
visualizes the data using a stacked bar chart to show the sales of each product category over time. The
program should prompt the user to enter the file path and display the chart.**


````python
import pandas as pd
import matplotlib.pyplot as plt

# Function to visualize sales data using a stacked bar chart
def visualize_sales(file_path):
    # Read the CSV file
    try:
        df = pd.read_csv(file_path)
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
        return
    except pd.errors.EmptyDataError:
        print("Error: The file is empty.")
        return

    # Ensure the DataFrame has the required columns (e.g., 'Date', 'Product Category', 'Sales')
    required_columns = ['Date', 'Product Category', 'Sales']
    if not all(col in df.columns for col in required_columns):
        print("Error: The required columns are missing in the CSV file.")
        return

    # Convert the 'Date' column to datetime type
    df['Date'] = pd.to_datetime(df['Date'])

    # Pivot the data to create a DataFrame suitable for a stacked bar chart
    pivot_df = df.pivot(index='Date', columns='Product Category', values='Sales').fillna(0)

    # Create a stacked bar chart
    pivot_df.plot(kind='bar', stacked=True, figsize=(12, 8))
    plt.title('Stacked Bar Chart of Sales by Product Category Over Time')
    plt.xlabel('Date')
    plt.ylabel('Sales')
    plt.legend(title='Product Category', bbox_to_anchor=(1.05, 1), loc='upper left')

    # Show the chart
    plt.show()

# Prompt the user to enter the file path
file_path = input("Enter the file path (CSV): ")

# Call the function to visualize sales data
visualize_sales(file_path)
````

### Write a Python program that reads the CSV file, calculates the mean, median, and mode of the test scores, and
### displays the results in a table.

#
**The program should do the following:**

- Prompt the user to enter the file path of the CSV file containing the student dataR
- Read the CSV file into a Pandas DataFrameR
- Calculate the mean, median, and mode of the test scores using Pandas toolsR
- Display the mean, median, and mode in a table.
######

#
Assume that the CSV file student_data.csv contains the following data:
|Student ID|Test Score|
|----------|----------|
1   |   85

4   |   75

The program should calculate the mean, median, and mode of the test scores and display the results
in a table.

### A11.

````python
import pandas as pd

# Function to calculate mean, median, and mode of test scores
def calculate_statistics(file_path):
    # Read the CSV file into a Pandas DataFrame
    try:
        df = pd.read_csv(file_path)
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
        return
    except pd.errors.EmptyDataError:
        print("Error: The file is empty.")
        return

    # Calculate mean, median, and mode
    mean_score = df['Test Score'].mean()
    median_score = df['Test Score'].median()
    mode_scores = df['Test Score'].mode()

    # Display the results in a table
    results_table = pd.DataFrame({
        'Statistic': ['Mean', 'Median', 'Mode'],
        'Value': [mean_score, median_score, ', '.join(map(str, mode_scores))]
    })

    print("\nResults:")
    print(results_table)

# Prompt the user to enter the file path of the CSV file
file_path = input("Enter the file path of the CSV file containing the student data: ")

# Call the function to calculate and display statistics
calculate_statistics(file_path)
````
