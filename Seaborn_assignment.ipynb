**Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.**

**Ans1:** \
**Seaborn is a powerful data visualization library in Python that is built on top of Matplotlib.\
It provides a high-level interface for creating attractive and informative statistical graphics.\
Here are five types of plots that we can create using Seaborn, along with their common use cases:**

 - These are just a few examples, and Seaborn provides a variety of other plots and customization options for creating informative and visually appealing visualizations.

1. **Scatter Plot:**
   - **Use Case:** Scatter plots are used to visualize the relationship between two continuous variables. They are helpful in identifying patterns, trends, and correlations in the data.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

 Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 3, 5]

# Create a scatter plot
sns.scatterplot(x=x, y=y)
plt.title('Scatter Plot Example')
plt.show()
```

2. **Bar Plot:**
   - **Use Case:** Bar plots are effective for comparing the values of different categories or groups. They are suitable for displaying categorical data and visualizing the distribution of values across categories.


```python

# Sample data
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

# Create a bar plot
sns.barplot(x=categories, y=values)
plt.title('Bar Plot Example')
plt.show()
```

3. **Box Plot:**
   - **Use Case:** Box plots (box-and-whisker plots) provide a visual summary of the distribution of a dataset. They display key statistical measures such as median, quartiles, and potential outliers.


```python
# Sample data
data = sns.load_dataset('tips')

# Create a box plot
sns.boxplot(x='day', y='total_bill', data=data)
plt.title('Box Plot Example')
plt.show()
```

4. **Violin Plot:**
   - **Use Case:** Violin plots combine aspects of box plots and kernel density estimation to provide a more detailed view of the distribution of data. They are useful for visualizing the distribution of a variable across different categories.


```python
# Sample data
data = sns.load_dataset('tips')

# Create a violin plot
sns.violinplot(x='day', y='total_bill', data=data)
plt.title('Violin Plot Example')
plt.show()

```


5. **Heatmap:**
   - **Use Case:** Heatmaps are used to visualize the magnitude of a relationship between two variables in a matrix format. They are commonly employed for exploring correlations in large datasets.


```python
# Sample data
data = sns.load_dataset('flights')

# Reshape data for heatmap
heatmap_data = data.pivot(index='month', columns='year', values='passengers')

# Create a heatmap
sns.heatmap(heatmap_data, cmap='viridis', annot=True, fmt='d')
plt.title('Heatmap Example')
plt.show()

```

**Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.**
- Note: timepoint, signal, event, and region are columns in the fmri dataset.

**Ans 2:**

```python
#loading the dataset fmri
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
#plotting a line plot for events and regions
sns.lineplot(x = "timepoint" , y = "signal", hue = "event", style = "region", data = fmri)

plt.title('Lineplot of fmri dataset')
plt.xlabel('timepoint')
plt.ylabel('signal')

plt.show()
```

**Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.**
 - Note: pclass, age, and fare are columns in the titanic dataset.
 
 **Ans 3:**

```python
#loading the dataset titanic
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# plotting boxplots using titanic dataset
plt.figure(figsize=(12,6))

# box plot for age
plt.subplot(1,2,1)
sns.boxplot(x = 'pclass',y = 'age',data = titanic)
plt.title('Box plot of age by pclass')

# box plot for fare
plt.subplot(1,2,2)
sns.boxplot(x = 'pclass', y = 'fare', data = titanic)
plt.title('Box plot of fare by pclass')

# for better spacing
plt.tight_layout()

plt.show()
```

**Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.**

**Ans 4:**

```python
# loading dataset diamonds
diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

```python
#plotting histogram with 'cut' as the hue
sns.histplot(data = diamonds , x = 'price' , hue = 'cut', bins = 30 , multiple= 'stack', palette = 'husl')

plt.title('Histogram of Price by cut')
plt.xlabel('Price')
plt.ylabel('Frequecy')
plt.show()
```
